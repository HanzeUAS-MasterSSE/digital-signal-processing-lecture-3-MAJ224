```python
import urllib.request
import librosa
import numpy as np
import matplotlib.pyplot as plt
from IPython import display

# Download and load the sound file
file_name = "2776.mp3"
url = f"https://bigsoundbank.com/UPLOAD/mp3/{file_name}"

urllib.request.urlretrieve(url, file_name)
sound_data, samplerate = librosa.load(file_name)
print(f"Sample rate: {samplerate} Hz")
print(f"Duration: {len(sound_data)/samplerate:.2f} seconds")

# Play the audio
display.display(display.Audio(data=sound_data, rate=samplerate))

# Plot the data in time domain
time = np.linspace(0, len(sound_data)/samplerate, len(sound_data))
plt.figure(figsize=(12, 4))
plt.plot(time, sound_data)
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Sound Signal in Time Domain')
plt.grid(True)
plt.show()
```

# Question 1

```python
from scipy.fft import fft, fftfreq

# Compute FFT
N = len(sound_data)
yf = fft(sound_data)
xf = fftfreq(N, 1/samplerate)

# Plot only the positive frequencies (first half of the spectrum)
plt.figure(figsize=(12, 4))
plt.plot(xf[:N//2], np.abs(yf[:N//2]))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Sound Signal in Frequency Domain (FFT)')
plt.xlim(1000, 2500)  # Focus on 1-2.5kHz range
plt.grid(True)
plt.show()
```

# Question 2

**Analysis of the Frequency Domain Plot:**

From the frequency domain plot, we can observe:

1. **Dominant Frequencies**: The plot shows which frequencies have the most energy/power in the signal. Peaks in the spectrum indicate the fundamental frequencies and harmonics present in the cockatiel sound.

2. **Frequency Range**: We can see the range of frequencies that make up the bird's call. Most bird vocalizations have significant energy in the 1-8 kHz range.

3. **Harmonic Structure**: Multiple peaks at different frequencies indicate harmonics (integer multiples of a fundamental frequency), which is typical of natural sounds and gives them their characteristic timbre.

4. **Spectral Distribution**: The distribution of energy across frequencies tells us about the sound's tonal quality - whether it's more pure-tone-like (narrow peaks) or noisy (broad spectrum).

5. **Bandwidth**: The spread of frequencies gives us information about the complexity of the sound - a wider bandwidth suggests a more complex, richer sound.

This frequency domain representation is crucial for:
- Audio filtering and processing
- Sound classification and recognition
- Understanding the acoustic properties of the sound
- Designing audio effects and modifications

```python
# Compute FFT for second sound
N_2 = len(sound_data_2)
yf_2 = fft(sound_data_2)
xf_2 = fftfreq(N_2, 1/samplerate_2)

# Compare the two frequency domain plots side by side
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

# First sound
ax1.plot(xf[:N//2], np.abs(yf[:N//2]))
ax1.set_xlabel('Frequency (Hz)')
ax1.set_ylabel('Magnitude')
ax1.set_title(f'Sound 1 ({file_name}) - Frequency Domain')
ax1.set_xlim(0, 5000)
ax1.grid(True)

# Second sound
ax2.plot(xf_2[:N_2//2], np.abs(yf_2[:N_2//2]))
ax2.set_xlabel('Frequency (Hz)')
ax2.set_ylabel('Magnitude')
ax2.set_title(f'Sound 2 ({file_name_2}) - Frequency Domain')
ax2.set_xlim(0, 5000)
ax2.grid(True)

plt.tight_layout()
plt.show()
```

**Comparison Analysis:**

Comparing the two frequency domain plots reveals:

1. **Different Spectral Signatures**: Each sound has a unique frequency distribution, acting like a "fingerprint" for that sound.

2. **Peak Locations**: The location of dominant peaks differs between the two sounds, indicating different fundamental frequencies and harmonic structures.

3. **Energy Distribution**: One sound might have more concentrated energy at specific frequencies (pure tones), while the other might have more distributed energy (noise-like).

4. **Bandwidth Differences**: The frequency range where significant energy is present can vary, telling us about the tonal complexity and richness of each sound.

5. **Spectral Envelope**: The overall shape of the spectrum is different, which contributes to how we perceive the sounds as being different.

These differences are why we can distinguish between different sounds - each has unique characteristics in the frequency domain that our ears and brains can detect.
